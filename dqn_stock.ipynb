```
import numpy as np
import pandas as pd


class Env:
  
  def __init__(self, target_code):
    self.target_code = target_code
    df = pd.read_csv('./000660.KS.csv')
    self.total_volume = 10000000
    self.episode_length = 200
    self.purchased = 0
    self.processed = 0
    
    df['p_close']  = df.Close.shift(1)
    df['r_open']   = df['Open']  / df['p_close']
    df['r_high']   = df['High']  / df['p_close']
    df['r_low']    = df['Low']   / df['p_close']
    df['r_close']  = df['Close'] / df['p_close']

    df['ma_5'] =  df.Close.rolling( 5).mean()
    df['ma_10'] = df.Close.rolling(10).mean()
    df['ma_20'] = df.Close.rolling(20).mean()

    df['r_ma_5']  = df['ma_5']   / df['p_close']
    df['r_ma_10'] = df['ma_10']  / df['p_close']
    df['r_ma_20'] = df['ma_20']  / df['p_close']
    df['r_vol']   = df['Volume'] / self.total_volume
    
    self.df = df.dropna()

  def reset(self):
    self.processed = 0
    self.current_pos = \
    np.random.randint(0, self.df.shape[0] - self.episode_length)
    #for i in range(self.episode_length):
    #  print(self.df.iloc[self.current_pos + i])
    
    
  def step(self, action):
    next_state = 0
    reward = 0
    done = False
    
    #if action == 0: # B
    #  self.purchased = self.df.iloc[self.current_pos]
    #else action == 1: # S  
    #  self.purchased = 0
    
    current_state = self.df.iloc[self.current_pos]
    

    if action == 0: # B
      print("action buy ")
      if self.purchased == 0:
        self.purchased = current_state['Close']
        print("buy at :", str(self.purchased))
      
    
    if action == 1: # S
      print("action sell")
      if self.purchased != 0:
        reward = (current_state['Close'] / self.purchased) 
        print("sell at : ", str(current_state['Close']), str(reward))
      self.purchased = 0

    if action == 2: # H
      print("action hold")
      reward = 0
      
    
    self.processed += 1
    self.current_pos += 1
    n_row = self.df.iloc[self.current_pos]
    next_state = [n_row['r_open'], n_row['r_high'], n_row['r_low']] # \
#                  n_row['r_close'], n_row['r_ma_5'], n_row['r_ma_10']]
#                  n_row['r_ma_20'], n_row['r_vol']
  
    
    # print("processed:" + str(self.processed) + " : action : " + str(action))
    if self.processed == self.episode_length:
      done = True

    return (next_state, reward, done)
   
    
env = Env('000660')
env.reset()

_done = False
while _done == False:
  a = np.random.randint(0, 3)
  next_state, reward, _done = env.step(a)
  print(str(a), next_state, reward) 
  
```
